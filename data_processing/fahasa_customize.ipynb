# Extract cột Series/Episode từ BookName

```python

def extract_series_episode(row):
    book_name = row['book_name']
    book_name_origi = row['book_name_origi']
    
    if " - Tập" in book_name:
        series, episode = book_name.split(" - Tập", 1)
        return pd.Series([series.strip(), f"Tập {episode.strip()}"])
    else:
        return pd.Series([book_name_origi, None])

# Apply the function to create 'series' and 'episode' columns
df[['series', 'episode']] = df.apply(extract_series_episode, axis=1)
```

# Extract cột Sell_method (tặng kèm hay không) và Gift_method (tặng kèm cái gì)

```python
# Function to determine sell_method based on "Tặng ..." in book_name
def determine_sell_method(book_name):
    if "Tặng" in book_name:
        return "Tặng " + book_name.split("Tặng", 1)[1].strip()
    else:
        return "Không tặng kèm"

# Apply the function to create the 'sell_method' column
df['sell_method'] = df['book_name'].apply(determine_sell_method)
```

```python
import pandas as pd
import re

# Function to remove 'đ', spaces, commas, and other unwanted characters, then return as a numeric value
def clean_currency(value):
    if isinstance(value, str):
        # Use regex to remove all non-numeric characters (except for the decimal point)
        cleaned_value = re.sub(r'[^\d.]', '', value)
        return pd.to_numeric(cleaned_value, errors='coerce')
    return value

# Apply the function to both 'new_price' and 'old_price' columns
df1['new_price'] = df1['new_price'].apply(clean_currency)
df1['old_price'] = df1['old_price'].apply(clean_currency)
df1['new_price'] = df1['new_price'] * 1000
df1['old_price'] = df1['old_price'] * 1000
```

## sửa cột quantity sold

```python
import pandas as pd

# Function to replace 'k', 'k+', and '+' with '000' or clean up the value
def replace_k(value):
    if isinstance(value, str):
        # Remove the '+' sign if it exists
        value = value.replace('+', '')
        
        # Check for 'k' or 'k+' and multiply by 1000
        if 'k' in value:
            return float(value.replace('k', '')) * 1000
    return value

# Apply the function to 'quantity_sold' column
df1['quantity_sold'] = df1['quantity_sold'].apply(replace_k)
```
